## **Naive Bayes**

**Step 1: Mengimpor the Libraries**

Seperti biasa, langkah pertamaya itu mengimpor pustaka NumPy, Pandas, dan Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Step 2: Mengimpor dataset**

Pada langkah ini, kita akan mengimpor dataset Bunga Iris yang disimpan di repositori github sebagai IrisDataset.csv dan menyimpannya ke dataset variabel. Setelah ini, kita tetapkan 4 variabel independen ke X dan variabel dependen 'spesies' ke Y. 5 baris pertama dari kumpulan data ditampilkan.

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Classification/master/IrisDataset.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

**Step 3: Split dataset menjadi data training dan data testing**

Setelah berhasil menampilkan data, kita harus membagi data menjadi data training dan data testing. Pada kumpulan data ini terdapat 150 baris dengan masing-masing 50 baris dari 3 kelas. Karena setiap kelas diberikan secara berurutan, kita perlu membagi dataset secara acak. Di sini, kita memiliki test_size=0.2, yang berarti bahwa 20% dari dataset akan digunakan untuk tujuan testing sebagai set testing dan 80% sisanya akan digunakan sebagai set training untuk melatih model klasifikasi Naive Bayes.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

**Step 4: Feature Scaling**

Dataset diperkecil ke rentang yang lebih kecil menggunakan opsi Feature Scalling. Dalam hal ini, nilai X_train dan X_test diperkecil ke nilai yang lebih kecil untuk meningkatkan kecepatan program.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Step 5: Training the Naive Bayes Classification model on the Training Set**

Pada langkah ini, kami memperkenalkan GaussianNB yang digunakan dari libraries sklearn.naive_bayes. Disini kita menggunakan model Gaussian, ada beberapa model lain seperti Bernoulli, Categorical dan Multinomial. Di sini, kami menetapkan kelas GaussianNB ke pengklasifikasi variabel dan menyesuaikan nilai X_train dan y_train untuk tujuan pelatihan.

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

**Step 6: Prediksi Dataset Testing**

Setelah model dilatih, kami menggunakan classifier.predict() untuk memprediksi nilai untuk Test set dan nilai yang diprediksi disimpan ke variabel y_pred.

```python
y_pred = classifier.predict(X_test) 
y_pred
```

**Step 7: Confusion Matrix dan Akurasi**

Ini adalah langkah yang banyak digunakan dalam teknik klasifikasi. Dalam hal ini, kita melihat Keakuratan model yang dilatih dan memplot Confusion matrix.

Confusion matrix adalah tabel yang digunakan untuk menunjukkan jumlah prediksi yang benar dan salah pada suatu masalah klasifikasi ketika nilai sebenarnya dari Test Set diketahui.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

Dari Confusion matrix di atas, kami menyimpulkan bahwa, dari 30 set data tes, 29 diklasifikasikan dengan benar dan hanya 1 yang salah diklasifikasikan. Ini memberi kami akurasi tinggi 96,67%.

**Step 8: Membandingkan nilai sebenarnya dan nilai hasil prediksi**

Pada langkah ini, Pandas DataFrame dibuat untuk membandingkan nilai terklasifikasi dari set Tes asli (y_test) dan hasil prediksi (y_pred).

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```
